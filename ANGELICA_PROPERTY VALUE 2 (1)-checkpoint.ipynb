### Feature Engineering

```python
## Preliminary check to see how many rows there are for sale_price
data.PropertyEvaluationvalue.count()
```

```python
## Change numeric variables to numeric data types
num_vars = ['NoOfResidentialUnits', 'NoOfCommercialUnits', 'TotalNoOfUnits', 
            'LandAreaInSqFt', 'GrossAreaInSqFt', 'PropertyEvaluationvalue']

for col in num_vars:
    data[col] = pd.to_numeric(data[col], errors = 'coerce')
#If coerce, then invalid parsing will be set as NaN.
#If ignore, then invalid parsing will return the input.    
```

```python
## Change categorical variables to the category data type
cat_vars = ['BldgClassCategory', 'TaxClass_AtEvaluationTime',
            'BldgClass_AtEvaluationTime', 'Address',
            'Borough', 'Block', 'Lot', 'ZipCode']

for col in cat_vars:
    data[col] = data[col].astype('O') #'O' means the values which are of object type
```

```python
## Find all rows where values are blank, a single space, or only a single hyphen
## Replace these values with NaN
data.replace(['', ' ', '-'], np.nan, inplace = True)
data.isnull().sum()
```

The sum of residential and commercial units should be equal to total units. I will remove any rows where there might be any abnormal values in the total units, including where there are 0 total units.

```python
data = data[data.TotalNoOfUnits  == data.NoOfResidentialUnits  + data.NoOfCommercialUnits ]
data = data[data.TotalNoOfUnits  != 0]
```

### Training and Evaluation

#### Dummyfication

```python
data = pd.get_dummies(data, columns = cat_vars, drop_first = True)
data.drop(['LandAreaInSqFt', 'TotalNoOfUnits'], axis = 1, inplace = True)
```

### Splitting the model into train and test
We are comparing the results of three algorithms

```python
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import mean_squared_error as MSE
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score

X = data.drop('PropertyEvaluationvalue', axis = 1)

y = data.PropertyEvaluationvalue

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 3)
```

```python
y_train.shape
```

```python
LR_model = LinearRegression()
DT_model = DecisionTreeRegressor()
RFR_model = RandomForestRegressor(n_estimators = 50)

model_list = [LR_model, DT_model, RFR_model]

for model in model_list:
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    print(str(model) + " test RMSE:", round(MSE(y_test, predictions)**0.5, 2))
    print(str(model) + " test corr:", np.corrcoef(y_test, predictions)[0,1])
    print(str(model) + " r2:", r2_score(y_test, predictions), "\n")
```

```python
from sklearn.metrics import mean_absolute_error as MAE
from sklearn.metrics import mean_absolute_percentage_error as MAPE

print("Mean Absolute Error: ", MAE(y_test, predictions))
print("Mean Absolute Percent Error: ", MAPE(y_test, predictions))
print("Mean Error: ", np.mean(predictions - y_test)) 
```

```python
fig = plt.figure(figsize = (10, 10))
plt.scatter(x = y_test, y = predictions)
m, b = np.polyfit(y_test, predictions, 1)
plt.plot(y_test, m*y_test+b, color='red')
```
