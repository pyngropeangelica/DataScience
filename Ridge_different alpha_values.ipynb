```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
```

```python
X,y = make_regression(n_samples=100, n_features=1, 
                      n_informative=1, n_targets=1,noise=20,random_state=13)
```

```python
plt.scatter(X,y)
```

For Linear Regression we compare the results with ridge

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X,y)
print(lr.coef_)
print(lr.intercept_)
```

```python
from sklearn.linear_model import Ridge
```

```python
rr = Ridge(alpha=10)
rr.fit(X,y)
print(rr.coef_)
print(rr.intercept_)
```

```python
rr1 = Ridge(alpha=100)
rr1.fit(X,y)
print(rr1.coef_)
print(rr1.intercept_)
```

As we can see the value of alpha or lambda highly impacts the value of m. We can see as we increase the value of alpha the value of m decreses significantly.

```python
plt.plot(X,y,'b.')
plt.plot(X,lr.predict(X),color='red',label='alpha=0')
plt.plot(X,rr.predict(X),color='green',label='alpha=10')
plt.plot(X,rr1.predict(X),color='orange',label='alpha=100')
plt.legend()
```

In this plot, the optimal value of alpha would be 10 i.e., the green line.

```python
def linear_regression(X,y,alpha=1):
    x_mean = X.mean()
    y_mean = y.mean()
    
    num = 0
    den = 0
    
    for i in range(X.shape[0]):
        num = num + (y[i] - y_mean) * (X[i] - x_mean)
        den = den + (X[i] - x_mean) * (X[i] - x_mean)
        
    m = num/(den + alpha)
    b = y_mean - m*x_mean
        
    return m,b
```
